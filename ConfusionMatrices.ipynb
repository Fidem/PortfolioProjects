#Confusion Matrices

A confusion matrix generally looks like this



|           |                             | Actual            | Actual                      |
|-----------|-----------------------------|-------------------|-----------------------------|
|           |                             | Has long toenails | Does not have long toenails |
| Predicted | Has long toenails           | True Positive     | False Positive              |
| Predicted | Does not have long toenails | False Negative    | True Negative               |

A confusion matrix essentially tells you what your machine learning algorithm did right and what it did wrong.


We can go deeper though, and look at some metrics that make use of confusion matrices to measure a models performance.

# Sensitivity and Specificity

Sensitivity will tell us what percentage of patients WITH long toenails were correctly identified.
Sensitivity is given by:
# $Sensitivity = \frac{True\: Positives}{True\: Positives + False\: Negatives}$

Specificity tells us what percentage of patients WITHOUT long toenails were correctly identified.
Specificity is given by:
# $Specificity = \frac{True\:Negatives}{True\:Negatives + False\:Positives}$

#ROC and AUC!!!

A ROC graph is a "Receiver Operator Characteristic" graph.  This is a plot of sensitivity (True positive rate) vs 1-specificity (False Positive Rate).
roc.png

Each point on the ROC graph represents a threshold, an the best one is the BEST possible point would be at (0,1), or as close it it as possible.

The AUC is the AREA UNDER CURVE of the ROC graph.  This makes it very easy to compare different ROC curves to another.  The higher AUC is the better one.

#Precision and Recall and F1 Score


Precision is the proportion of positive results that were correctly classified
# $Precision = \frac{True\:Positives}{True\:Positives + False\:Positives}$
Recall is the Sensitivity
# $Recall = \frac{True\:Positives}{True\:Positives+False\:Negatives}$
F1 Score is the harmonic mean of the precision and the recall.
# $F_1 = 2\cdot\frac{Precision \cdot Recall}{Precision+Recall}$


# Type I and Type II Errors

A False positive is known as a Type I Error.  This is when you predict something to be true when it is actually false.

A False Negative is known as a Type II Error.  This is when you predict something to be false when it is actually true.
